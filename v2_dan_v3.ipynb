```python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘  DSC MCF ITB 2026 â€” Team Tensor Selow                                  â•‘
# â•‘  Improved Solution â€” Target Score < 5                                  â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# â”€â”€ 0. INSTALL & IMPORT â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
!pip install xgboost lightgbm statsmodels --quiet

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import make_scorer
import xgboost as xgb
import lightgbm as lgb
from statsmodels.tsa.holtwinters import ExponentialSmoothing

plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['font.size'] = 11
sns.set_style('whitegrid')
print('âœ… Libraries loaded')

# â”€â”€ 1. LOAD DATA â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
from google.colab import drive
drive.mount('/content/drive')

BASE_PATH = '/content/drive/MyDrive/MCF/'  # â† sesuaikan

df_klaim = pd.read_csv(BASE_PATH + 'Data_Klaim.csv')
df_polis = pd.read_csv(BASE_PATH + 'Data_Polis.csv')
df_sample = pd.read_csv(BASE_PATH + 'sample_submission.csv')
print(f'Klaim: {df_klaim.shape} | Polis: {df_polis.shape}')

# â”€â”€ 2. PREPROCESSING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
klaim = df_klaim.copy()
klaim.columns = [
    'claim_id','nomor_polis','reimburse_cashless','inpatient_outpatient',
    'icd_diagnosis','icd_description','status_klaim',
    'tgl_bayar','tgl_masuk','tgl_keluar',
    'nominal_disetujui','nominal_rs','lokasi_rs'
]
for col in ['tgl_bayar','tgl_masuk','tgl_keluar']:
    klaim[col] = pd.to_datetime(klaim[col], errors='coerce')
klaim['nominal_disetujui'] = pd.to_numeric(klaim['nominal_disetujui'], errors='coerce')
klaim['nominal_rs']        = pd.to_numeric(klaim['nominal_rs'], errors='coerce')

# Filter valid paid claims
klaim_paid = klaim[
    (klaim['status_klaim'] == 'PAID') &
    (klaim['nominal_disetujui'] > 0) &
    (klaim['tgl_masuk'].notna())
].copy()

print(f'Valid PAID claims: {len(klaim_paid)}')

# Feature engineering - individual level
klaim_paid['year']    = klaim_paid['tgl_masuk'].dt.year
klaim_paid['month']   = klaim_paid['tgl_masuk'].dt.month
klaim_paid['los']     = (klaim_paid['tgl_keluar'] - klaim_paid['tgl_masuk']).dt.days.clip(lower=0)
klaim_paid['icd_chapter'] = klaim_paid['icd_diagnosis'].fillna('Z').str[0]
klaim_paid['processing_days'] = (klaim_paid['tgl_bayar'] - klaim_paid['tgl_masuk']).dt.days

# Polis preprocessing
polis = df_polis.copy()
polis.columns = ['nomor_polis','plan_code','gender','tgl_lahir','tgl_efektif','domisili']
polis['tgl_lahir']   = pd.to_datetime(polis['tgl_lahir'], format='%Y%m%d', errors='coerce')
polis['tgl_efektif'] = pd.to_datetime(polis['tgl_efektif'], errors='coerce')
ref_date = pd.Timestamp('2025-01-01')
polis['usia'] = ((ref_date - polis['tgl_lahir']).dt.days / 365.25).round(1)
polis['age_group'] = pd.cut(polis['usia'], bins=[0,18,30,45,60,200],
                              labels=['<18','18-30','31-45','46-60','60+'])
polis['tenure_years'] = ((ref_date - polis['tgl_efektif']).dt.days / 365.25).round(1)

df = klaim_paid.merge(polis, on='nomor_polis', how='left')
print(f'Merged: {df.shape}')

# â”€â”€ 3. EDA â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
monthly_raw = df.groupby(['year','month']).agg(
    freq=('claim_id','count'),
    total=('nominal_disetujui','sum')
).reset_index()
monthly_raw['severity'] = monthly_raw['total'] / monthly_raw['freq']
monthly_raw['year_month_dt'] = pd.to_datetime(monthly_raw[['year','month']].assign(day=1))
monthly_raw = monthly_raw.sort_values('year_month_dt').reset_index(drop=True)
print('\nðŸ“Š Monthly Data (all 19 months):')
print(monthly_raw[['year','month','freq','severity','total']].to_string())

# Plot time series
fig, axes = plt.subplots(3, 1, figsize=(14, 12))
for ax, (col, label, scale) in zip(axes, [
    ('freq','Claim Frequency',1),
    ('severity','Avg Severity (IDR Juta)',1e6),
    ('total','Total Claim (IDR Miliar)',1e9)
]):
    ax.plot(monthly_raw['year_month_dt'], monthly_raw[col]/scale,
            'o-', linewidth=2, markersize=6)
    ax.axvline(pd.Timestamp('2025-01-01'), color='red', linestyle='--',
               alpha=0.6, label='2025 Start')
    ax.set_title(label, fontweight='bold')
    ax.legend(); ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=30)
plt.tight_layout()
plt.savefig('monthly_trends.png', dpi=150, bbox_inches='tight')
plt.show()

# EDA plots
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes[0,0].hist(np.log1p(df['nominal_disetujui']), bins=50, color='steelblue', edgecolor='white')
axes[0,0].set_title('Distribusi Log Nominal Klaim'); axes[0,0].set_xlabel('log(Nominal)')

gender_agg = df.groupby('gender')['nominal_disetujui'].mean()
axes[0,1].bar(gender_agg.index, gender_agg.values/1e6, color=['steelblue','coral'])
axes[0,1].set_title('Avg Klaim per Gender'); axes[0,1].set_ylabel('IDR Juta')

plan_agg = df.groupby('plan_code')['nominal_disetujui'].mean()
axes[0,2].bar(plan_agg.index, plan_agg.values/1e6, color=['gold','tomato','seagreen'])
axes[0,2].set_title('Avg Klaim per Plan Code'); axes[0,2].set_ylabel('IDR Juta')

ip_agg = df.groupby('inpatient_outpatient')['nominal_disetujui'].mean()
axes[1,0].bar(ip_agg.index, ip_agg.values/1e6, color='mediumpurple')
axes[1,0].set_title('Avg Klaim: IP vs OP'); axes[1,0].set_ylabel('IDR Juta')

month_freq = df.groupby('month')['claim_id'].count()
axes[1,1].bar(month_freq.index, month_freq.values, color='teal')
axes[1,1].set_title('Frekuensi per Bulan (Seasonality)'); axes[1,1].set_xticks(range(1,13))

top_icd = df.groupby('icd_chapter')['nominal_disetujui'].sum().nlargest(10)
axes[1,2].barh(top_icd.index, top_icd.values/1e9, color='darkorange')
axes[1,2].set_title('Total Klaim per ICD Chapter'); axes[1,2].set_xlabel('IDR Miliar')
plt.tight_layout()
plt.savefig('eda_plots.png', dpi=150, bbox_inches='tight')
plt.show()

# Heatmap
num_cols = ['nominal_disetujui','nominal_rs','los','processing_days','usia']
corr = df[num_cols].dropna().corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn', center=0, square=True)
plt.title('Correlation Heatmap', fontweight='bold')
plt.tight_layout(); plt.savefig('heatmap.png', dpi=150, bbox_inches='tight'); plt.show()

# Boxplot age group
plt.figure(figsize=(10,5))
df_box = df[df['nominal_disetujui'] < df['nominal_disetujui'].quantile(0.99)]
df_box.boxplot(column='nominal_disetujui', by='age_group', figsize=(10,5))
plt.suptitle(''); plt.title('Distribusi Klaim per Age Group')
plt.ylabel('Nominal Klaim (IDR)'); plt.xlabel('Age Group')
plt.tight_layout(); plt.savefig('boxplot_age.png', dpi=150, bbox_inches='tight'); plt.show()

print('\nRata-rata klaim per inpatient/outpatient:')
print(df.groupby('inpatient_outpatient')['nominal_disetujui'].agg(['count','mean','median']).round(0))
print('\nRata-rata klaim per plan code:')
print(df.groupby('plan_code')['nominal_disetujui'].agg(['count','mean','median']).round(0))
print('\nTop ICD Chapter by total klaim:')
print(df.groupby('icd_chapter')['nominal_disetujui'].agg(['count','mean','sum'])
        .sort_values('sum', ascending=False).head(10).round(0))

# YoY Analysis
print('\nâ”€â”€ YoY Analysis Janâ€“Jul 2024 vs 2025 â”€â”€')
yoy_comp = {}
for yr in [2024, 2025]:
    sub = df[(df['year']==yr) & (df['month'].isin(range(1,8)))]
    yoy_comp[yr] = {'freq': len(sub), 'total': sub['nominal_disetujui'].sum(),
                    'severity': sub['nominal_disetujui'].mean()}
    print(f"{yr}: freq={yoy_comp[yr]['freq']}, total={yoy_comp[yr]['total']/1e9:.2f}B, "
          f"severity={yoy_comp[yr]['severity']/1e6:.2f}M")
for metric in ['freq','total','severity']:
    r = yoy_comp[2025][metric] / yoy_comp[2024][metric]
    print(f"  YoY {metric}: {r:.4f} ({(r-1)*100:+.1f}%)")

# â”€â”€ 4. FEATURE IMPORTANCE (Individual Level) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df_model = df.copy()
le = LabelEncoder()
cat_cols = ['reimburse_cashless','inpatient_outpatient','icd_chapter',
            'lokasi_rs','plan_code','gender','domisili','age_group']
for col in cat_cols:
    df_model[col] = le.fit_transform(df_model[col].astype('object').fillna('UNK'))
feat_cols = cat_cols + ['month','year','los','usia','tenure_years']
df_clean = df_model[feat_cols + ['nominal_disetujui']].dropna()
rf_indiv = RandomForestRegressor(n_estimators=200, max_depth=8, random_state=42, n_jobs=-1)
rf_indiv.fit(df_clean[feat_cols], df_clean['nominal_disetujui'])
fi_indiv = pd.Series(rf_indiv.feature_importances_, index=feat_cols).sort_values(ascending=False)
plt.figure(figsize=(9,5))
fi_indiv.plot(kind='bar', color='teal', edgecolor='white')
plt.title('Feature Importance â€” RF (Individual Claim Level)', fontweight='bold')
plt.ylabel('Importance'); plt.xticks(rotation=45)
plt.tight_layout(); plt.savefig('feature_importance_individual.png', dpi=150, bbox_inches='tight'); plt.show()
print('\nðŸ“Œ Feature Importance (Individual):')
print(fi_indiv.round(4).to_string())

# â”€â”€ 5. MAPE HELPER â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def mape_score(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask]))

# â”€â”€ 6. TIME-SERIES MODELING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Kita punya 19 bulan data. Masalah sebelumnya: lag12 butuh 12 bulan lookback
# sehingga hanya tersisa 7 baris setelah dropna.
# SOLUSI: Buat feature set yang tidak bergantung lag12, sehingga semua 19 bulan bisa dipakai.

m = monthly_raw.copy()  # 19 baris, Jan2024-Jul2025

# Time features
m['t']           = np.arange(len(m))
m['month_sin']   = np.sin(2 * np.pi * m['month'] / 12)
m['month_cos']   = np.cos(2 * np.pi * m['month'] / 12)
m['is_2025']     = (m['year'] == 2025).astype(int)
m['quarter']     = ((m['month'] - 1) // 3 + 1)

# Lag features (hanya lag 1, 2, 3 â€” tidak ada lag12)
for lag in [1, 2, 3]:
    m[f'freq_lag{lag}']     = m['freq'].shift(lag)
    m[f'severity_lag{lag}'] = m['severity'].shift(lag)
    m[f'total_lag{lag}']    = m['total'].shift(lag)

# Rolling mean (window 3, shift 1 agar tidak bocor)
m['freq_roll3']     = m['freq'].shift(1).rolling(3, min_periods=1).mean()
m['severity_roll3'] = m['severity'].shift(1).rolling(3, min_periods=1).mean()
m['total_roll3']    = m['total'].shift(1).rolling(3, min_periods=1).mean()

# Setelah dropna untuk lag3 saja, tersisa 16 baris (Apr2024 â€“ Jul2025)
m_model = m.dropna(subset=['freq_lag3','severity_lag3','total_lag3']).reset_index(drop=True)
print(f'\nâœ… Training rows tersedia: {len(m_model)} (sebelumnya hanya 7!)')
print(m_model[['year','month','freq','severity','total']].to_string())

FEATURES = [
    't', 'month', 'month_sin', 'month_cos', 'is_2025', 'quarter',
    'freq_lag1','freq_lag2','freq_lag3','freq_roll3',
    'severity_lag1','severity_lag2','severity_lag3','severity_roll3',
    'total_lag1','total_lag2','total_lag3','total_roll3',
]
TARGETS = ['freq', 'severity', 'total']

# â”€â”€ 7. CROSS VALIDATION (TimeSeriesSplit) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Cross validation untuk time series: TIDAK boleh shuffle, urutan waktu harus terjaga.
# TimeSeriesSplit: train fold selalu di kiri, val fold di kanan secara kronologis.

print('\nâ”€â”€ Cross Validation (TimeSeriesSplit, n_splits=4) â”€â”€')
tscv = TimeSeriesSplit(n_splits=4, test_size=2)  # val = 2 bulan tiap fold

cv_results = {}

def cv_model(model, X, y, cv):
    mapes = []
    for train_idx, val_idx in cv.split(X):
        X_tr, X_v = X.iloc[train_idx], X.iloc[val_idx]
        y_tr, y_v = y.iloc[train_idx], y.iloc[val_idx]
        model.fit(X_tr, y_tr)
        pred = np.maximum(model.predict(X_v), 1)
        mapes.append(mape_score(y_v.values, pred))
    return np.mean(mapes), np.std(mapes)

X_all = m_model[FEATURES]

for target in TARGETS:
    y_all = m_model[target]
    cv_results[target] = {}

    # XGBoost
    xgb_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=2,
                                   subsample=0.8, random_state=42, verbosity=0)
    mean_m, std_m = cv_model(xgb_model, X_all, y_all, tscv)
    cv_results[target]['XGBoost'] = (mean_m, std_m)
    print(f'XGBoost    | {target:10s} | CV MAPE: {mean_m:.4f} Â± {std_m:.4f}')

    # LightGBM
    lgb_model = lgb.LGBMRegressor(n_estimators=100, learning_rate=0.1, num_leaves=8,
                                    min_child_samples=3, random_state=42, verbosity=-1)
    mean_m, std_m = cv_model(lgb_model, X_all, y_all, tscv)
    cv_results[target]['LightGBM'] = (mean_m, std_m)
    print(f'LightGBM   | {target:10s} | CV MAPE: {mean_m:.4f} Â± {std_m:.4f}')

    # Random Forest
    rf_model = RandomForestRegressor(n_estimators=100, max_depth=4,
                                      min_samples_leaf=2, random_state=42, n_jobs=-1)
    mean_m, std_m = cv_model(rf_model, X_all, y_all, tscv)
    cv_results[target]['RandomForest'] = (mean_m, std_m)
    print(f'RandomForest| {target:10s} | CV MAPE: {mean_m:.4f} Â± {std_m:.4f}')

# â”€â”€ 8. HYPERPARAMETER TUNING (GridSearchCV dengan TimeSeriesSplit) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print('\nâ”€â”€ Hyperparameter Tuning â”€â”€')
# Karena data kecil, kita tuning XGBoost & LightGBM dengan parameter kritis saja

tscv_tune = TimeSeriesSplit(n_splits=3, test_size=2)

neg_mape_scorer = make_scorer(
    lambda y_true, y_pred: -mape_score(y_true, np.maximum(y_pred, 1)),
    greater_is_better=True
)

xgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth':    [2, 3],
    'learning_rate':[0.05, 0.1],
    'subsample':    [0.7, 1.0],
}

best_xgb_params = {}
for target in TARGETS:
    y_all = m_model[target]
    gs = GridSearchCV(
        xgb.XGBRegressor(random_state=42, verbosity=0),
        xgb_param_grid, cv=tscv_tune, scoring=neg_mape_scorer,
        n_jobs=-1, refit=True
    )
    gs.fit(X_all, y_all)
    best_xgb_params[target] = gs.best_params_
    print(f'XGBoost best params [{target}]: {gs.best_params_} | Score: {-gs.best_score_:.4f}')

lgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'num_leaves':   [4, 8, 16],
    'learning_rate':[0.05, 0.1],
    'min_child_samples': [2, 3],
}

best_lgb_params = {}
for target in TARGETS:
    y_all = m_model[target]
    gs_lgb = GridSearchCV(
        lgb.LGBMRegressor(random_state=42, verbosity=-1),
        lgb_param_grid, cv=tscv_tune, scoring=neg_mape_scorer,
        n_jobs=-1, refit=True
    )
    gs_lgb.fit(X_all, y_all)
    best_lgb_params[target] = gs_lgb.best_params_
    print(f'LightGBM best params [{target}]: {gs_lgb.best_params_} | Score: {-gs_lgb.best_score_:.4f}')

# â”€â”€ 9. FINAL MODELS (TRAIN ON ALL DATA) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print('\nâ”€â”€ Training Final Models on All 16 Months â”€â”€')

# â”€â”€ Buat future features (Agustus â€“ Desember 2025) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
future_months_dt = pd.date_range('2025-08-01', periods=5, freq='MS')
future_df = pd.DataFrame({
    'year_month_dt': future_months_dt,
    'year':  future_months_dt.year,
    'month': future_months_dt.month,
})
future_df['t']           = len(m) + np.arange(len(future_df))
future_df['month_sin']   = np.sin(2 * np.pi * future_df['month'] / 12)
future_df['month_cos']   = np.cos(2 * np.pi * future_df['month'] / 12)
future_df['is_2025']     = 1
future_df['quarter']     = ((future_df['month'] - 1) // 3 + 1)

# Fill lags dari data aktual (iteratively untuk akurasi lebih baik)
known_freq     = list(m['freq'].values)
known_severity = list(m['severity'].values)
known_total    = list(m['total'].values)

for i in range(5):
    row = {}
    for lag in [1,2,3]:
        src_idx = -(lag - i)
        f_val = known_freq[src_idx]     if src_idx < 0 and abs(src_idx) <= len(known_freq) else known_freq[-1]
        s_val = known_severity[src_idx] if src_idx < 0 and abs(src_idx) <= len(known_severity) else known_severity[-1]
        t_val = known_total[src_idx]    if src_idx < 0 and abs(src_idx) <= len(known_total) else known_total[-1]
        future_df.loc[i, f'freq_lag{lag}']     = f_val
        future_df.loc[i, f'severity_lag{lag}'] = s_val
        future_df.loc[i, f'total_lag{lag}']    = t_val

    future_df.loc[i, 'freq_roll3']     = np.mean(known_freq[-3:])
    future_df.loc[i, 'severity_roll3'] = np.mean(known_severity[-3:])
    future_df.loc[i, 'total_roll3']    = np.mean(known_total[-3:])

X_future = future_df[FEATURES]

# Train & predict dengan tuned models
final_preds = {'XGBoost':{}, 'LightGBM':{}, 'RandomForest':{}}

for target in TARGETS:
    y_all = m_model[target]

    # XGBoost tuned
    xgb_tuned = xgb.XGBRegressor(**best_xgb_params[target], random_state=42, verbosity=0)
    xgb_tuned.fit(X_all, y_all)
    final_preds['XGBoost'][target] = np.maximum(xgb_tuned.predict(X_future), 1)

    # LightGBM tuned
    lgb_tuned = lgb.LGBMRegressor(**best_lgb_params[target], random_state=42, verbosity=-1)
    lgb_tuned.fit(X_all, y_all)
    final_preds['LightGBM'][target] = np.maximum(lgb_tuned.predict(X_future), 1)

    # RandomForest (no tuning, sudah cukup baik)
    rf_final = RandomForestRegressor(n_estimators=200, max_depth=4,
                                      min_samples_leaf=2, random_state=42, n_jobs=-1)
    rf_final.fit(X_all, y_all)
    final_preds['RandomForest'][target] = np.maximum(rf_final.predict(X_future), 1)

# â”€â”€ 10. HOLT-WINTERS (Time Series Klasik) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print('\nâ”€â”€ Holt-Winters â”€â”€')
hw_preds = {}
# Gunakan semua 19 bulan data mentah untuk HW
for target in TARGETS:
    series = m[target].values  # 19 bulan
    # Coba dengan seasonal (butuh minimal 2x period, kita punya 19 > 12*1.5)
    try:
        hw = ExponentialSmoothing(series, trend='add', seasonal='add',
                                   seasonal_periods=12, initialization_method='estimated')
        hw_fit = hw.fit(optimized=True, use_brute=True)
        preds = hw_fit.forecast(5)
        print(f'  HW seasonal OK [{target}]: {np.round(preds, 0)}')
    except Exception as e:
        # Fallback: tanpa seasonal
        hw = ExponentialSmoothing(series, trend='add', initialization_method='estimated')
        hw_fit = hw.fit(optimized=True)
        preds = hw_fit.forecast(5)
        print(f'  HW no-seasonal [{target}]: {np.round(preds, 0)}')
    hw_preds[target] = np.maximum(preds, 1)

final_preds['HoltWinters'] = hw_preds

# â”€â”€ 11. YoY BASELINE (ANCHOR TERKUAT) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print('\nâ”€â”€ YoY Baseline â”€â”€')
# Hitung YoY ratio dari 7 bulan pertama
data_2024 = df[df['year']==2024]
data_2025 = df[df['year']==2025]

def yoy_ratio(col_2024, col_2025):
    return col_2025 / col_2024

freq_24  = len(data_2024[data_2024['month'].isin(range(1,8))])
freq_25  = len(data_2025[data_2025['month'].isin(range(1,8))])
total_24 = data_2024[data_2024['month'].isin(range(1,8))]['nominal_disetujui'].sum()
total_25 = data_2025[data_2025['month'].isin(range(1,8))]['nominal_disetujui'].sum()
sev_24   = total_24 / freq_24
sev_25   = total_25 / freq_25

ratio_freq  = freq_25  / freq_24
ratio_total = total_25 / total_24
ratio_sev   = sev_25   / sev_24

print(f'YoY ratio freq={ratio_freq:.4f}, sev={ratio_sev:.4f}, total={ratio_total:.4f}')

# 2024 Augâ€“Dec actuals
aug_dec_2024 = m[(m['year']==2024) & (m['month']>=8)].sort_values('month')
yoy_preds_dict = {
    'freq':     aug_dec_2024['freq'].values     * ratio_freq,
    'severity': aug_dec_2024['severity'].values * ratio_sev,
    'total':    aug_dec_2024['total'].values    * ratio_total,
}
final_preds['YoY'] = yoy_preds_dict

for target in TARGETS:
    print(f'  YoY pred [{target}]: {np.round(final_preds["YoY"][target], 0)}')

# â”€â”€ 12. ENSEMBLE (INVERSE-CV-MAPE WEIGHTED + YoY ANCHOR) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print('\nâ”€â”€ Building Ensemble â”€â”€')

# Evaluasi val MAPE terakhir (last 3 months sebagai holdout)
holdout_n = 3
train_cv = m_model.iloc[:-holdout_n]
val_cv   = m_model.iloc[-holdout_n:]
X_tr_cv = train_cv[FEATURES]
X_v_cv  = val_cv[FEATURES]

holdout_mape = {'XGBoost':{}, 'LightGBM':{}, 'RandomForest':{}, 'HoltWinters':{}, 'YoY':{}}
print('\nHoldout MAPE (last 3 months):')

for target in TARGETS:
    y_tr = train_cv[target]
    y_v  = val_cv[target]

    # XGBoost
    m_xgb = xgb.XGBRegressor(**best_xgb_params[target], random_state=42, verbosity=0)
    m_xgb.fit(X_tr_cv, y_tr)
    holdout_mape['XGBoost'][target] = mape_score(y_v.values, np.maximum(m_xgb.predict(X_v_cv), 1))

    # LightGBM
    m_lgb = lgb.LGBMRegressor(**best_lgb_params[target], random_state=42, verbosity=-1)
    m_lgb.fit(X_tr_cv, y_tr)
    holdout_mape['LightGBM'][target] = mape_score(y_v.values, np.maximum(m_lgb.predict(X_v_cv), 1))

    # RandomForest
    m_rf = RandomForestRegressor(n_estimators=200, max_depth=4,
                                   min_samples_leaf=2, random_state=42, n_jobs=-1)
    m_rf.fit(X_tr_cv, y_tr)
    holdout_mape['RandomForest'][target] = mape_score(y_v.values, np.maximum(m_rf.predict(X_v_cv), 1))

    # HoltWinters (train on train_cv, forecast holdout_n + 5)
    try:
        hw_tmp = ExponentialSmoothing(train_cv[target].values, trend='add',
                                       seasonal='add', seasonal_periods=12,
                                       initialization_method='estimated')
        hw_tmp_fit = hw_tmp.fit(optimized=True)
        hw_v_pred = hw_tmp_fit.forecast(holdout_n)
    except:
        hw_tmp = ExponentialSmoothing(train_cv[target].values, trend='add',
                                       initialization_method='estimated')
        hw_tmp_fit = hw_tmp.fit(optimized=True)
        hw_v_pred = hw_tmp_fit.forecast(holdout_n)
    holdout_mape['HoltWinters'][target] = mape_score(y_v.values, np.maximum(hw_v_pred, 1))

    # YoY holdout pred: val bulan-bulan adalah Mei, Jun, Jul 2025
    # Cari padanan 2024
    val_months = val_cv['month'].values
    yoy_v_preds = []
    for vm in val_months:
        row_2024 = m[(m['year']==2024) & (m['month']==vm)]
        if len(row_2024) > 0:
            base_val = row_2024[target].values[0]
            r = ratio_freq if target=='freq' else (ratio_sev if target=='severity' else ratio_total)
            yoy_v_preds.append(base_val * r)
        else:
            yoy_v_preds.append(y_v.mean())
    holdout_mape['YoY'][target] = mape_score(y_v.values, np.array(yoy_v_preds))

    for model_name in holdout_mape:
        print(f'  {model_name:15s} | {target:10s} | MAPE: {holdout_mape[model_name][target]:.4f}')

# Hitung ensemble weights berdasarkan holdout MAPE (inverse)
print('\nâ”€â”€ Final Ensemble Predictions â”€â”€')
ensemble_final = {}
weight_log = {}

for target in TARGETS:
    mape_vals = {m_name: holdout_mape[m_name][target] for m_name in holdout_mape}
    # Inverse MAPE weight + YoY mendapat floor weight minimum 25%
    raw_weights = {m_name: 1.0/(v+1e-6) for m_name, v in mape_vals.items()}
    total_w = sum(raw_weights.values())
    norm_weights = {m_name: w/total_w for m_name, w in raw_weights.items()}

    # Ensure YoY has at least 25% weight (kuat secara empiris untuk asuransi)
    if norm_weights['YoY'] < 0.25:
        deficit = 0.25 - norm_weights['YoY']
        norm_weights['YoY'] = 0.25
        other_total = sum(v for k,v in norm_weights.items() if k != 'YoY')
        for k in norm_weights:
            if k != 'YoY':
                norm_weights[k] -= deficit * (norm_weights[k]/other_total)

    weight_log[target] = norm_weights
    print(f'\n{target} weights: { {k: round(v,3) for k,v in norm_weights.items()} }')

    ensemble_pred = np.zeros(5)
    for m_name, w in norm_weights.items():
        p = final_preds[m_name][target]
        ensemble_pred += w * np.array(p)
    ensemble_final[target] = np.maximum(ensemble_pred, 1)
    print(f'{target} predictions: {np.round(ensemble_final[target], 0)}')

# â”€â”€ 13. VISUALISASI PREDIKSI â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, axes = plt.subplots(3, 1, figsize=(14, 13))
for ax, (target, label, scale, unit) in zip(axes, [
    ('freq',     'Claim Frequency',       1,    'Count'),
    ('severity', 'Claim Severity',        1e6,  'IDR Juta'),
    ('total',    'Total Claim',           1e9,  'IDR Miliar'),
]):
    ax.plot(m['year_month_dt'], m[target]/scale, 'o-',
            color='steelblue', linewidth=2, label='Historical')
    ax.plot(future_months_dt, yoy_preds_dict[target]/scale, 's--',
            color='orange', alpha=0.85, linewidth=1.5, label='YoY Pred')
    ax.plot(future_months_dt, ensemble_final[target]/scale, 'D-',
            color='crimson', linewidth=2.5, markersize=8, label='Ensemble Pred')
    ax.axvspan(future_months_dt[0], future_months_dt[-1], alpha=0.07, color='yellow')
    ax.axvline(pd.Timestamp('2025-08-01'), color='gray', linestyle=':', alpha=0.6)
    ax.set_title(f'{label} â€” Historical + Forecast (Augâ€“Dec 2025)', fontweight='bold')
    ax.set_ylabel(unit); ax.legend(); ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=30)
plt.tight_layout()
plt.savefig('forecast_final.png', dpi=150, bbox_inches='tight')
plt.show()

# Model comparison bar chart
holdout_df = pd.DataFrame(holdout_mape).T
holdout_df.columns = TARGETS
holdout_df.plot(kind='bar', figsize=(11,5), colormap='Set2', edgecolor='white')
plt.title('Holdout MAPE per Model & Target', fontweight='bold')
plt.ylabel('MAPE'); plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig('model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

# Feature importance XGBoost (freq)
xgb_fi = xgb.XGBRegressor(**best_xgb_params['freq'], random_state=42, verbosity=0)
xgb_fi.fit(X_all, m_model['freq'])
fi = pd.Series(xgb_fi.feature_importances_, index=FEATURES).sort_values(ascending=True).tail(15)
plt.figure(figsize=(9,6))
fi.plot(kind='barh', color='steelblue')
plt.title('Feature Importance â€” XGBoost (Frequency)', fontweight='bold')
plt.xlabel('Importance')
plt.tight_layout(); plt.savefig('feature_importance_ts.png', dpi=150, bbox_inches='tight'); plt.show()

# â”€â”€ 14. SUBMISSION FILE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
submission_rows = []
for i, dt in enumerate(future_months_dt):
    ym = dt.strftime('%Y_%m')
    submission_rows.append({'id': f'{ym}_Claim_Frequency', 'value': round(float(ensemble_final['freq'][i]),     4)})
    submission_rows.append({'id': f'{ym}_Claim_Severity',  'value': round(float(ensemble_final['severity'][i]), 4)})
    submission_rows.append({'id': f'{ym}_Total_Claim',     'value': round(float(ensemble_final['total'][i]),    4)})

submission = pd.DataFrame(submission_rows)
print('\nðŸ“‹ Final Submission:')
print(submission.to_string(index=False))

SAVE_PATH = BASE_PATH + 'submission_v2.csv'
submission.to_csv(SAVE_PATH, index=False)
print(f'\nâœ… Saved to {SAVE_PATH}')
print(f'âœ… Total rows: {len(submission)} (expected 15)')

# â”€â”€ 15. PROYEKSI 2026 â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print('\nâ”€â”€ Proyeksi 2026 â”€â”€')
actual_2025 = m[m['year']==2025][['month','freq','severity','total']].copy()
pred_aug_dec_df = pd.DataFrame({
    'month': range(8,13),
    'freq':     ensemble_final['freq'],
    'severity': ensemble_final['severity'],
    'total':    ensemble_final['total'],
})
full_2025 = pd.concat([actual_2025, pred_aug_dec_df]).reset_index(drop=True)
full_2025['year'] = 2025

proj_2026 = full_2025.copy()
proj_2026['year']     = 2026
proj_2026['freq']     = full_2025['freq']     * ratio_freq
proj_2026['severity'] = full_2025['severity'] * ratio_sev
proj_2026['total']    = full_2025['total']    * ratio_total

print(proj_2026[['month','freq','severity','total']].round(0).to_string(index=False))
total_2026 = proj_2026['total'].sum()
total_2025 = full_2025['total'].sum()
print(f'\nðŸ“Š 2026 Total Klaim  : {total_2026/1e9:.2f} IDR Miliar')
print(f'ðŸ“Š 2025 Total Klaim  : {total_2025/1e9:.2f} IDR Miliar')
print(f'ðŸ“Š Growth YoY 26/25  : {(total_2026/total_2025-1)*100:+.1f}%')
print(f'ðŸ“Š Avg Severity 2026 : {proj_2026["total"].sum()/proj_2026["freq"].sum()/1e6:.2f} IDR Juta')

# Plot proyeksi 2026
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
for ax, (col, label, scale, unit) in zip(axes, [
    ('freq','Claim Frequency',1,'Count'),
    ('severity','Avg Severity',1e6,'IDR Juta'),
    ('total','Total Claim',1e9,'IDR Miliar'),
]):
    ax.plot(range(1,13), full_2025[col]/scale, 'o-', label='2025', color='steelblue', linewidth=2)
    ax.plot(range(1,13), proj_2026[col]/scale, 's--', label='2026 Proj.', color='crimson', linewidth=2)
    ax.set_title(label, fontweight='bold'); ax.set_xlabel('Month'); ax.set_ylabel(unit)
    ax.legend(); ax.set_xticks(range(1,13)); ax.grid(True, alpha=0.3)
plt.suptitle('Proyeksi 2026 vs 2025 â€” Portfolio Level', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('projection_2026.png', dpi=150, bbox_inches='tight')
plt.show()

# Rekomendasi bisnis
print("""
â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
ðŸ“‹ REKOMENDASI BISNIS
â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

ðŸŽ¯ 1. Risk Selection
- ICD Chapter C (Kanker) & I (Jantung/Kardiovaskular) mendominasi total klaim.
  â†’ Perketat underwriting & medical check-up untuk usia 46+ dengan riwayat
    penyakit kronis. Pertimbangkan exclusion atau waiting period lebih panjang.
- Plan M-001 (Worldwide) severitas 37% lebih tinggi dari M-003 (Domestik).
  â†’ Penyesuaian premi atau naikkan deductible untuk Plan M-001.

ðŸ¥ 2. Pencegahan & Deteksi Dini
- Length of Stay (LOS) adalah faktor #1 penentu besar klaim (importance 38%).
  â†’ Implementasi case management untuk rawat inap > 5 hari.
- Lokasi RS (lokal vs internasional) adalah faktor #2 (importance 29%).
  â†’ Pertimbangkan steering program ke RS rekanan dengan tarif ternegosiasi.
- Usia adalah faktor #3 â€” nasabah 60+ berkontribusi klaim terbesar.
  â†’ Program wellness & preventive care untuk segmen senior.

ðŸ“ˆ 3. Proyeksi & Aksi Premium
- Proyeksi 2026: total klaim diestimasi ~{:.1f}% dibanding 2025.
- Inflasi medis + trend klaim â†’ rekomendasikan kenaikan premi 8â€“12%.
- Bulan Januari & Februari secara historis memiliki frekuensi tinggi.
  â†’ Pastikan reserve klaim mencukupi di awal tahun.
â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
""".format((total_2026/total_2025 - 1) * 100))
```
